```python
import pandas as pd
import numpy as np

print("Loading Men's and Women's Stage 2 Predictions...")
men_preds = pd.read_csv('mens_stage2_predictions.csv')
women_preds = pd.read_csv('womens_stage2_predictions.csv')

# Combine them into one dataframe
final_submission = pd.concat([men_preds, women_preds])

# Enforce Kaggle's Log Loss Safety Constraints
# If an absolute massive upset occurs, a 1.0 or 0.0 prediction triggers an infinite penalty.
final_submission['Pred'] = np.clip(final_submission['Pred'], 0.0001, 0.9999)

# Save the official Kaggle file
final_submission.to_csv('submission.csv', index=False)

print("\nSUCCESS! 'submission.csv' is generated!")
print(f"Total Matchups Predicted: {len(final_submission)} (Ready for Kaggle Upload)")
```
